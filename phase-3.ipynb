```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data Structures :[restaurant_id,2021,2022,2023,2024]
sales_data=np.array([
    [1,50000,40000,60000,55000],
    [2,20000,65000,78000,52000],
    [3,70000,60000,50000,45000],
    [4,90000,100000,60000,505000],
    [5,45000,90000,100000,15000],
])

print("==== Zomato Restro Analysis ====")
print("\n Sale Data Shape:",sales_data.shape)
print("Restro 1st three:\n",sales_data[:3])
```

```python
#total sales per year
print(np.sum(sales_data[:,1])) #single column
print(np.sum(sales_data,axis=0))
print(np.sum(sales_data[:,1:],axis=0))

#minimum sales per restro
min_sales=np.min(sales_data[:,1:],axis=1)   # Axis 1 goes for single row and axis 0 goes for single column
print(min_sales)

#maximum sales per year
max_sales=np.max(sales_data[:,1:],axis=0)
print(max_sales)

#Avg sales per restaurant
avg_sales=np.average(sales_data[:,1:],axis=1)
print(avg_sales)
```

```python
cumulative_sum=np.cumsum(sales_data[:,1:],axis=1)
print(cumulative_sum)

plt.figure(figsize=(5,4))
plt.plot(np.mean(cumulative_sum,axis=0))
plt.title("Average cumulative sales across all restaurants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
